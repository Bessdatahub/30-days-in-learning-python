##### Simple Arithmetic Operations
Write a program that takes two numbers as input and prints their sum, difference, multiplication, and division.

```python
number_1= int(input("Write down a number: "))

number_2 = int(input("Write down a number: "))

Adition= number_1 + number_2

Difference= number_2 - number_1

Multiplication= number_2 * number_1

Division= number_2 / number_1

print(f"The sum of these two numbers are {Adition} \nThe diff of these two numbers are {Difference} \nThe product of these two numbers are {Multiplication} \nThe division of these two numbers are {Division}")
```
