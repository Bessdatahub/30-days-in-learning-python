##### 5. Write a program that takes two numbers from the user and prints whether the first number is greater than, less than, or equal to the second number.

```python
A= int(input("Write down the first number:"))
B = int(input("Write down the second number: "))

if A > B:
    print(f"The first number {A} is greater than {B}")
elif B > A:
    print(f"The second number {B} is greater than {A}")

else:
    print(f"They are both the same")
```

##### 7. Write a Python program that takes two numbers from the user and prints the result of their division. Make sure to handle the case when the second number is zero.

```python
C= float(input("Write down a number: "))
D= float(input("Write down a number: "))

if D == 0:
    print("Division by zero is not possible")
else:
    result= C/D
    print(f"The result is {result}")
```

##### 8.Write a Python program that converts Celsius to Fahrenheit. Formula: F = (C * 9/5) + 32

```python
Celsius= float(input("Write down the temperature in celsius:"))

Fahrenheit= (Celsius * 9/5) + 32

print(f"The temperature in Fahrenheit is {Fahrenheit}")
```

##### Write a program that takes a string from the user and prints whether it is a palindrome (the same forwards and backwards).

```python
word =input("Write down a word:")

if word == word[::-1]:
    print(f"it is a palindrome {word}")
else:
    print("It is not a palindrome")
```
