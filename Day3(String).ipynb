##### 4.Slicing in Python

When it comes to slicing, there is the start, end and step
Note that for end you add +1 to it else python does not include the end, Python is 0 index from the front and -1 indexing from the back

Square Brackets are used when slicing

```python
# This start with 0 index, stops at the last digit and in 2 step
products = "Laptop Table"

products[0::2]
```

```python
# This start with 3 index, stops at the last digit 11 and in 3 step
products[3:11:3]
```

```python
# this will bring empty because its start with -5 and going back to -9 wont be possible indexing is on a straight line
products[-5:-9]
```

```python
# Here python recognises -1+1 as 0 and 0 index is at the front meanwhile the starting is -5 so it cannot go back
products[-5:-1+1]
```

```python
products[-5:-1]
```
